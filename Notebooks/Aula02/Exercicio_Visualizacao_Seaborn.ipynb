```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('../../Datasets/german_credit_data.csv')
data
```

```python
# Gráfico de Linha: Variação de idades ao longo dos registros
plt.figure(figsize=(10, 6))
sns.lineplot(data=data['Age'])
plt.title('Variação de Idades')
plt.xlabel('Registro')
plt.ylabel('Idade')
plt.show()
```

```python
# Gráfico de Barras: Contagem de Pessoas por Sexo

plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Sex')
plt.title('Contagem de Pessoas por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de Barras (Countplot) usando barplot
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Sex', y='Age', estimator=len) 
# O parâmetro estimator=len especifica que queremos contar o número de ocorrências em vez de calcular a média.
plt.title('Contagem de Pessoas por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.show()
```

```python
# Gráfico de Barras (Barplot) usando barplot
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Sex', y='Credit amount', ci=None)
# O parâmetro ci=None desativa a exibição dos intervalos de confiança.
plt.title('Média do Valor do Crédito por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Média do Valor do Crédito')
plt.show()
```

```python
# Histograma: Distribuição de Idades
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de Dispersão: Idade vs. Valor do Crédito
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Credit amount', alpha=0.5)
plt.title('Gráfico de Dispersão: Idade vs. Valor do Crédito')
plt.xlabel('Idade')
plt.ylabel('Valor do Crédito')
plt.show()
```

```python
# Gráfico de Boxplot: Distribuição de Idade por Sexo
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Sex', y='Age')
plt.title('Comparação de Idade por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Idade')
plt.show()
```

```python
# Calcular a matriz de correlação
correlation_matrix = data.corr()

# Gráfico de Mapa de Calor (Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Mapa de Calor: Matriz de Correlação')
plt.show()
```
